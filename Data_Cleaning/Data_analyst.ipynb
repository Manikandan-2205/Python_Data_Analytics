# String Value Manipulation

```python
df.select_dtypes('string').describe().T
```

```python
(
    df
    .select_dtypes('string')
    .memory_usage(deep=True)
    .sum()
)
```

```python
(
    df
    .select_dtypes('string')
    .astype('category')
    .memory_usage(deep=True)
    .sum()
)
```

```python
(
    df
    .isna()
    .mean()
    .mul(100)
    .pipe(lambda ser: ser[ser > 0])
)
```

```python
(
    df
    .select_dtypes('string')
    .isna()
    .mean()
    .mul(100)
    .pipe(lambda ser: ser[ser > 0])
)
```
